Starting with the energy balance in Table 2-3 in Deen, we can cross out terms related to time, to more than one direction, and to internal heating terms to get the following equation:

$$ \alpha*\frac{d^2T}{dx^2} = 0 $$ where $$ \alpha = \frac{k}{\rho*C_p} $$

Thus,

$$ \frac{d}{dx}(k_1 + B(T-T_1)) \frac{dT}{dx} = 0 $$

In the above equation, T refers to temperature, Beta is a given constant, and k1 is the thermal conductivity at point one from the problem statement in the book. There are only boundary conditions in this case, but the solution of the problem should be stable as long as beta isn't 0.

The above equation can be integrated first to get a constant on the right side of the equation. The remainder is then integrated by parts as shown below.

$$ \int k_1 + BT - BT_1 dT = \int c dx $$
$$ k_1T + \frac{B}{2}T^2 - BT_1T = cx + c $$
$$ \frac{B}{2}T^2 + T*(k_1-BT_1) = c_1x+c_2 $$

Using the initial conditions of $T(0) = T_1$ and $T(l) = T_2$, solve for $c_1$ and $c_2$.

$$ c_1 = \frac{1}{l}[0.5*B*(T_2^2-T_1^2) + (k_1 - BT_1)(T_2-T_1)] $$

$$ c_2 = \frac{BT_1^2}{2} + (k_1-BT_1)T_1 $$

Plug in $c_1$ and $c_2$ and then solve for T using the quadratic formula.

$$T-T_1 = \frac{-k_1}{B} + \frac{1}{B} \sqrt{k_1^2 - 2k_1BT_1 + B^2T_1^2 + 2B(c_1x+c_2)} $$

In this equation,

$$ c_1x + c_2 = \frac{x}{l}[0.5*B*(T_2^2-T_1^2) + (k_1 - BT_1)(T_2-T_1)] + \frac{BT_1^2}{2} + (k_1-BT_1)T_1 $$

Simplify using... 

$$ k_2 = k_1 + B(T_2 - T_1) $$ 

$$ k_m = \frac{2k_1 + B(T_2 - T_1)}{2} $$

This leads to a final equation of:

$$ T - T_1 = \frac{1}{B} \sqrt{k_1^2 + 2Bk_m(T_2-T_1) \frac{x}{l}} - 1 $$
$$ k_m = 0.5(k_1 + k_2) $$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1000,100)
k_1 = 0.1
k_2 = 15
k_m = 0.5*((k_1)+(k_2))
T1 = 0
T2 = 100
B = 5
l = 1000
T = (1/B)*np.sqrt(k_1**2+2*B*k_m*(T2-T1)*(x/l)) - 1 + T1
plt.plot(x,T)
plt.title("Variation in k_1")
plt.xlabel("x")
plt.ylabel("T")
```
