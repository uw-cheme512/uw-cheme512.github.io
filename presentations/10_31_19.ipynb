```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ti = 100
Tf = 200
Bi = 1
Ri = 1
Ro = 10
fig, ax = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
k = 1

for Bi in np.logspace(-3, 3, 7):
    c1 = (-Ti+Tf)/(-np.log(Ri/Ro)+(1/Bi))
    c2 = Ti - c1*np.log(Ri)

    r = np.linspace(Ri+0.001, Ro-0.001, 101)
    T = c1*np.log(r) + c2
    q = -k*c1/r
    qtot = -k*c1*np.log(r/Ri)

    ax.plot(r, T, label='Bi = {}'.format(Bi))
    ax.set_title('T')
    ax2.plot(r, q, label='Bi = {}'.format(Bi))
    ax2.set_title('q')
    ax3.plot(r, qtot, label='Bi = {}'.format(Bi))
    ax3.set_title('qtot as a function of Ro')
    
    
ax.legend()
ax2.legend()
```
