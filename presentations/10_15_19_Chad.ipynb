```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R1 = 0.1
R2 = 0.105
r = np.linspace(R1,R2,101)


fig, ax = plt.subplots(figsize=(6,4))
Bii = 10
Bio = 10
for Bio in [0.01, 0.1, 1, 10, 10000000]:
    H = Bio/Bii

    Ti = 320
    To = 300

    c2 = (H*To*(1 - Bii*np.log(R1)) + Ti*(1 + Bio*np.log(R2)))/(1 + H + Bio*np.log(R2) - H*Bii*np.log(R1))
    c1 = Bio*(To - c2)/(1 + Bio*np.log(R2))

    T = c1*np.log(r) + c2
    ax.plot(r, T, label='Bio = {}'.format(Bio))

ax.legend()
```
