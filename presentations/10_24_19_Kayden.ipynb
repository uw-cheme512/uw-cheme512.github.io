In this problem, it's steady state, no velocity, and no heat generation. Moreover, we only consider the r direction. Hence we can write the unsolved equation in this form.

$$0= \alpha[\frac{1}{r^2}\frac{\partial}{\partial r}(r^2\frac{\partial T}{\partial r})]$$

B.C.1 $$q=k\frac{dT}{dr}=-hi [T-T(r)]_{r=r_i}$$

B.C.2 $$q=k\frac{dT}{dr}=h_0 [T-T(r)]_{r=r_0}$$

From the equation above

$$\frac{\partial}{\partial r}(r^2\frac{\partial T}{\partial r})=0$$

$$r^2\frac{\partial T}{\partial r}=C_1$$

$$\frac{\partial T}{\partial r}= \frac{C_1}{r^2}$$

$$T=-\frac{C_1}{r}+C_2$$

We have the general solution for the problem, and we'll plug in the B.C. 1 for getting the relation of $C_1$ and $C_2$

$$\frac{\partial T}{\partial r}=-\frac{h_i}{k}[T-T(r)]_{r=r_i}$$

$$=-\frac{h_i}{k}[(-\frac{C_1}{r_i}+C_2)-T_i$$

$$\frac{\partial T}{\partial r}=\frac{h_i}{kr_i}C_1-\frac{h_i}{k}C_2+\frac {h_i}{k}T_i$$

plug in B.C.2

$$\frac{\partial T}{\partial r}=\frac {h_0}{k}[(-\frac{C_1}{r_0}+C_2)-T_0]$$

$$=-\frac{h_0}{kr_0}C_1+\frac{h_0}{k}C_2-\frac{h_0}{k}T_0$$

compare these two equation, we can get

$$\frac{h_i}{kr_i}C_1-\frac{h_i}{k}C_2+\frac{h_i}{k}T_i=-\frac{h_0}{kr_0}C_1+\frac{h_0}{k}C_2-\frac{h_0}{k}T_0$$

then we can solve $C_1$ and $C_2$ from this equation

$$C_1[\frac{hi}{kr_i}+\frac{h_0}{kr_0}]-C_2[\frac{h_0+h_i}{k}]+(\frac{h_0+h_i}{k})T_0=0$$

$$T_i=-\frac{c_1}{r_i}+C_2$$

$$T_0=-\frac{C_1}{r_0}+C_2$$

$$T_i-T_0= -\frac{C_1}{ri}+\frac{C_1}{r_0}\rightarrow \frac{r_i-r_0}{r_0r_i}C_1=T_i-T_0$$

$$C_1= T_i-T_0* \frac{r_0r_i}{r_i-r_0}$$

plug in $C_1$ in to the former equation and find the value of $C_2$

$$[(T_i-T_0)*\frac{r_0r_i}{r_i-r_0}][\frac{h_i}{kr_i}+\frac{h_0}{kr_0}]-c_2(\frac{h_0+hi}{k})+ (\frac{h_0+h_i}{k})T_0=0$$

$$C_2=[(T_i-T_0)*(\frac{r_0r_i}{r_i-r_0})(\frac{h_i}{kr_i}+\frac{h_0}{kr_0})+\frac{h_0+h_i}{k}T_0]*\frac{k}{h_0+h_i}$$

$$C_2=\frac{(T_i-T_0)(r_0r_i)}{r_i-r_0}(\frac{r_0h_i+r_ih_0}{kr_0r_i})*\frac{k}{h_0+h_i})+T_0$$

plug in $C_1$ and $C_2$ into Temp. euqation

$$T=-\frac{(T_i-T_0)r_0r_i}{(r_i-r_0)}\frac{1}{r}+\frac{(T_i-T_0)(r_0r_i)}{r_i-r_0} \frac{r_0h_i+r_ih_0}{kr_0r_i}*\frac{k}{(h_0+h_i)}+T_0$$


$$\frac{T-T_0}{T_i-T_0}=\frac{(r_0/r)-1}{(r_0/r_i)-1+(r_0/r_i)(k/r_ih_i)+k/r_0h_0}$$

$$Q=-k\frac{dt}{dr}$$

$$Q=-k\frac{dT}{dr}=-h_0(T-T_i)|_{r=r_0}$$
