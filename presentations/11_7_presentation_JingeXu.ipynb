## Create homogeneous BC
This problem has non-homogeneous boundary conditions, kind of difficult to handle, so we want to convert them into homogeneous boundary conditions.
Consider when t $\rightarrow \infty$, the system achieves equilibrium, the temperature distribution is independent on time, only dependent on location(in this problem, x),we define a function $u_e(x)$as the temperature distribution in the infinite plate when t$\rightarrow \infty$,viz,$$u_e(x)=\lim_{t\rightarrow \infty} u(x,t)$$

Rewrite the boundary conditions which equilibrium temperature distribution would satisfy
$$
u_e(-L)=T_0\\
u_e(L)=T_0\\
\frac{du_e(0)}{dx}=0
$$
we can solve $u_e(x)$
$$u_e(x)=T_0$$

Now we define $$v(x,t)=u(x,t)-u_e(x)$$
and derivatives
$$
\frac{\partial u}{\partial t}=\frac{\partial v}{\partial t}+\frac{\partial u_e}{\partial t}=\frac{\partial v}{\partial t}\\
\frac{\partial^2 u}{\partial x^2}=\frac{\partial^2 v}{\partial x^2}+\frac{\partial^2 u_e}{\partial x^2}=\frac{\partial^2 v}{\partial x^2}
$$

For v(x,t), the initial condition is
$$
v(x,0)=u(x,0)-u_e(x)=f(x)-T_0
$$
Boundary conditions are
$$
v(-L,t)=u(-L,t)-u_e(-L)=T_0-T_0=0\\
v(L,t)=u(L,t)-u_e(L)=T_0-T_0=0\\
$$
Now we successfully have homogeneous boundary conditions! Next we could solve for v(x,t)!

## separate variables to solve PDE
So the PDE is
$$
\frac{\partial v}{\partial t}=k\frac{\partial^2 v}{\partial x^2}
$$
with boudary conditions and initial conditions
$$
v(-L,t)=0\\
v(L,t)=0\\
v(x,0)=(T_c-T_0)cos(\frac{\pi x}{2L}),-L<x<+L
$$
We also notice that two boundary conditions are symmetric, so v(x,t) would satisfy
$$
\frac{\partial v(0)}{\partial x}=0
$$

To solve v(x,t), we assume $v(x,t)=\phi(x)G(t)$
so we can rewrite the PDE as
$$
\phi(x)\frac{dG}{dt}=kG(t)\frac{d^2\phi}{d x^2}
$$
Then
$$
\frac{1}{kG(t)} \frac{dG}{dt}=\frac{1}{\phi(x)} \frac{d^2\phi}{dx^2}\\
v(-L,t)=\phi(-L)G(t)=0\\
v(L,t)=\phi(-L)G(t)=0\\
$$
because we don't expect G(t)=0,so the boundary conditions are
$$
\phi(-L)=0\\
\phi(L)=0\\
$$
since two sides are derivatives to different variables, to satisfy the PDE for all x and t, two sides of the equation must be the same constant, and we set the constant to be $-\lambda$

### express $\phi(x)$
Now we need to find the valid values for $\lambda$, we try to solve right side:
$$
\frac{\partial^2\phi}{\partial x^2}+\lambda \phi(x)=0
$$
eigenvalues
$$
r=\pm \sqrt{-\lambda}
$$

when $\lambda>0$,
$$
\phi(x)=c_1cos(\sqrt\lambda x)+c_2sin(\sqrt\lambda x)
$$
and first derivative
$$
\frac{d\phi(x)}{dx}=-c_1\sqrt\lambda sin(\sqrt\lambda x)+c_2\sqrt\lambda cos(\sqrt\lambda x)
$$
since the first derivative satisfy 
$$
\frac{d\phi(0)}{dx}=0\\
c_2=0\\
\phi(x)=c_1cos(\sqrt\lambda x)$$
Apply the boundary condition
$$\phi(x)=c_1cos(\sqrt\lambda L)=0$$
If $c_1=0$,it will be trivial solution,which we don't expect,so
$$
cos(\sqrt\lambda L)=0\\
\sqrt\lambda L=(n+\frac{1}{2})\pi ,n=0,1,2,3,,,,\\
\lambda=[(n+\frac{1}{2})\frac{\pi}{L}]^2, ,n=0,1,2,3,,,,
$$
For arbitrary $c_1$, we assume it 1,there is
$$\phi(x)=cos[(n+\frac{1}{2})\frac{\pi}{L}x]\\
where \ \lambda=[(n+\frac{1}{2})\frac{\pi}{L}]^2, \ n=0,1,2,3,,,,,,$$


when $\lambda=0$,
$$
\phi(x)=c_1+c_2x\\
$$
Apply the boudary conditions:
$$c_1=0,c_2=0$$
That's a trivial solution,
so $\lambda=0$ is not a valid eigenvalue

when $\lambda<0$
$$
\phi(x)=c_1cosh(\sqrt-\lambda x)+c_2sinh(\sqrt-\lambda x)\\
$$
and the first derivative
$$
\sqrt-\lambda[-c_1sinh(\sqrt-\lambda x)+c_2cosh(\sqrt-\lambda x)\\
$$
Apply$\frac{d\phi(0)}{dx}=0$,
$$
c_2=0
$$
so 
$$
\phi(x)=c_1cosh(\sqrt-\lambda x)
$$
Then apply the boundary condition
$$
\phi(L)=c_1cosh(\sqrt-\lambda L)=0
$$
because $coshx>0$,
$$c_1=0$$
so when $\lambda<0$,there are no valid eigenvalues

Therefore, we can solve the right side of the equation
$$\phi(x)=cos[(n+\frac{1}{2})\frac{\pi}{L}x], n=0,1,2,3,,,,,
$$

### Express G(t)
Now we consider the left side of the equation
$$
\frac{1}{kG(t)} \frac{dG}{dt}=-\lambda \\
G(t)=\sum_{n=0}^\infty B_n exp[-k((n+\frac{1}{2})\frac{\pi}{L})^2t], n=0,1,2,3,,,,,,,,
$$

Now we can express $v(x,t)=\phi(x)G(t)$
$$
v(x,t)=\sum_{n=0}^\infty B_n cos[(n+\frac{1}{2})\frac{\pi}{L}x] exp[-k((n+\frac{1}{2})\frac{\pi}{L})^2t],\ \ n=0,1,2,3,,,,
$$

plug in the initail condition 
$$
v(x,0)=\sum_{n=0}^\infty B_n cos[(n+\frac{1}{2})\frac{\pi}{L}x]=(T_c-T_0)cos(\frac{\pi x}{2L}),-L<x<+L
$$

therefore, $$n=0,\  B_0=T_c-T_0$$
and we get the expression for $v(x,t)$
$$
v(x,t)=(T_c-T_0)cos[\frac{\pi}{2L}x] exp[-k(\frac{\pi}{2L})^2t] ,  -L<x<+L
$$
and the expression for $u(x,t)$
$$
u(x,t)=T_0+(T_c-T_0)cos[\frac{\pi}{2L}x] exp[-k(\frac{\pi}{2L})^2t]  , -L<x<+L
$$
this is the temperature distribution in the infinite plate

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

```python
n = 0
L = 10
k = 0.01
tc=200
t0=25
d = 101
x = np.linspace(-L, L, d)
t = np.linspace(0, 15000, d)
u = np.zeros((x.shape[0], t.shape[0]))
xv, tv = np.meshgrid(x, t)

# Initial values
B = tc-t0
u = t0+B * np.cos(np.pi*xv/2/L)*np.exp(-k*tv*(np.pi/2/L)**2)

for ni in range(2, 1000):
    B = tc-t0
    u = t0+B * np.cos(np.pi*xv/2/L)*np.exp(-k*tv*(np.pi/2/L)**2)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xv, tv, u, cmap=cm.coolwarm)
ax.view_init(azim=50,elev=20)

ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('u')
ax.set_zlim(20, 210)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

for i in range(0, d):
    ax.plot(xv[i, :], u[i, :], c='k', alpha=i/d)
ax.set_xlabel('x')
ax.set_ylabel('u')
```

```python
fig, ax = plt.subplots(figsize=(8,8))

for i in range(0,d):
    ax.plot(tv[:, i], u[:, i], c='k', alpha=(1 - (i/d)))
ax.set_xlabel('t')
ax.set_ylabel('u')
```
