# Analytical solution
To tackle down this problem, we can make several assumptions to make it an one-dimensional problem.

1. $T=T(x,y,z)$, but as $B<<L$ and $B<<W$, the temperature can be seen as a function of z alone.
2. neglect the heat loss at the end (area $2BW$) and at the edges (area $2*2BL$)$\Rightarrow$ heat out only happens on two $WL$ area.
3. heat transfer coefficient $h$ and thermal conductivity $k$ are independent of temperature and position.

Shell balance gives:
$$2BWq_z|_z-2BWq_z|_{z+ \Delta z}-h(2W\Delta z)(T-T_a)=0$$

Where the first two terms stand for the heat flux of conduction within the fin, and the last term stands for the heat loss from fin to the air through conduction.

Dividing by $2BW\Delta z$ and taking $\Delta z \to 0$ gives:
$$-\frac{dq_z}{dz}=\frac{h}{B}(T-T_a)$$

Fourier's law gives that the heat flux can be written as(z-direction only):
$$q_z=-k\frac{dT}{dz}$$

Substituting $q_z$, we can get a second-order ODE:
$$\frac{d^2T}{dz^2}=\frac{h}{kB}(T-T_a)$$

With the following boundary conditions:
$$B.C.1:z=0, T=T_w$$
$$B.C.2:z=L, \frac{dT}{dz}=0$$

To simplify the solution, we introduce some dimensionless quantities:
$$\phi=\frac{T-T_a}{T_w-T_a}$$
$$\zeta=\frac{z}{L}$$
$$N^2=\frac{hL^2}{kB}$$

The problem then takes the form:
$$\frac{d^2\phi}{d\zeta^2}=N^2\zeta$$

Where the boundary conditions are:
$$\phi|_{\zeta=0}=1, \frac{d\phi}{d\zeta}|_{\zeta=1}=0$$

Since the equation is a homogeneous second-order ODE with constant coefficient, it can be easily solved with characteristic equation:
$$\phi=C_1e^{N\zeta}+C_2e^{-N\zeta}=acosh(N\zeta)+bsinh(N\zeta)$$

Applying the boundary conditions to solve for constants:
$$B.C.1:1=a*1\Rightarrow a=1$$
$$B.C.2:0=N*1[sinh(N*1)+bcosh(N*1)]\Rightarrow b=-\frac{sinh(N)}{cosh(N)}=-tanh(N)$$

The final solution is:
$$\phi=cosh(N\zeta)-tanh(N)sinh(N\zeta)$$

Which can be rearranged:
$$\phi=\frac{cosh[N(1-\zeta)]}{cosh(N)}$$

# Physical meaning
The dimensionless term N can also be rearranged into a function of Biot number:
$$B^2*N^2=\frac{hL^2}{kB}*B^2=\frac{hL^2B}{k}=Bi*L^2\Rightarrow N^2=Bi*(\frac{L}{B})^2$$

Where in this case Biot number can be written as:
$$Bi=\frac{L_ch}{k}$$

In which $L_c=$ characteristic length $=\frac{V_{body}}{A_{surface}}$

Since the surface that we account for heat loss are only two $WL$ side, $L_c$ should be:
$$L_c=\frac{V_{body}}{A_{surface}}=\frac{2BWL}{2WL}=B$$

Thus, Biot number for this problem is:
$$Bi=\frac{hB}{k}$$

For the assumption we already assume that B<<L, we can set it to a reasonable ratio so that $\frac{L}{B}$ is large enough to validate the assumption. Set it to different values and see how temperature profile of dimensionless temperature $\phi$ change with different Biot numbers.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
zeta=np.linspace(0, 1, 1001)
Biot=np.array([0.0001, 0.001, 0.01, 0.1, 1, 5, 10, 25, 50, 100])
phi=0
LB=100

for i, Bi in enumerate(Biot):
    N=(LB*Bi)**0.5
    phi=np.cosh(N*(1-zeta))/np.cosh(N)
    plt.plot(zeta, phi, label='Bi='+str(Bi))
    plt.legend()
    plt.xlabel('$\zeta$')
    plt.ylabel('$\phi$')
    plt.title('L/B='+str(LB))
```
