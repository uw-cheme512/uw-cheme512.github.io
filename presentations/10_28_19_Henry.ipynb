## Solution of the differential equation using analytical method

By conservation equation of energyï¼š

$$\rho \hat{C_p}\frac{DT}{Dt}=-\nabla \cdot q+H_v$$

from fourier's law

$$q=-k\nabla T$$

Since k is not constant here, we shouldn't combine them as $k\nabla^2T$ 

$$\Rightarrow \rho \hat{C_p}\frac{DT}{Dt}=-\nabla \cdot (-k\nabla T)+H_v$$

Modify the spherical form of the energy conversation equation in Deen Table 2-3 (You could also find it from page 14 to page 16 of Heat Transfer Solution by James H. VanSant), and it would be written as equation (1)ï¼š

$$\rho \hat{C_p}[\frac{\partial T}{\partial t}+v_r\frac{\partial T}{\partial r}+\frac{v_\theta}{r}\frac{\partial T}{\partial {\theta}}+\frac{v_\phi}{rsin{\theta}}\frac{\partial T}{\partial \phi}]= \frac{1}{r^2}\frac{\partial}{\partial r}(r^2 k\frac{\partial T}{\partial r})+\frac{1}{r^2sin\theta}\frac{\partial}{\partial\theta}(ksin\theta\frac{\partial T}{\partial \theta})+\frac{1}{r^2sin^2\theta}k\frac{\partial^2T}{\partial\phi^2}+H_v\text{------(1)}$$

Assuming system is steady state, there are no flow velocity and heat generation, and temperature only varies with radius.

Equation (1) could be reduced as belowï¼š

$$ \frac{1}{r^2}\frac{\partial}{\partial r}(r^2k\frac{\partial T}{\partial r})=0$$

Then, plug in $k=k_0+\beta(T-T_o)$

$$r^2(k_0+\beta(T-T_o))\frac{\partial T}{\partial r}=C_1$$

Solve differential equation by separationï¼š

$$\int (k_0+\beta(T-T_o)dT=\int \frac{C_1}{r^2}dr$$

$$\frac{\beta}{2}T^2+(k_0-\beta T_o)T=-\frac{C_1}{r}+C_2 \space\space \text{------(2)}$$

Now, using boundary conditions, solve for constants C1 and C2

$$T=T_i \space\space\space at\space\space\space r=r_i\space\space,\space\space\space\space T=T_o \space\space\space at\space\space\space r=r_o$$ 

$$\begin{cases}
\frac{\beta}{2}T_i^2+(k_0-\beta T_o)T_i=-\frac{C_1}{r_i}+C_2 &\text{------(3)} \\
\frac{\beta}{2}T_o^2+(k_0-\beta T_o)T_o=-\frac{C_1}{r_o}+C_2 &\text{------(4)}
\end{cases}$$

$$(3)-(4)$$

$$\Rightarrow \frac{\beta}{2}(T_i^2-T_o^2)+(k_0-\beta T_o)(T_i-T_o)=C_1(\frac{1}{r_o}-\frac{1}{r_i})$$

$$\Rightarrow (T_i-T_o)(\frac{\beta}{2}(T_i+T_o)+k_0-\beta T_o)=\frac{r_i-r_o}{r_or_i}C_1$$

$$\Rightarrow C_1=\frac{r_or_i}{r_i-r_o}(T_i-T_o)(\frac{\beta}{2}(T_i-T_o)+k_0)$$

$$C_2=\frac{\beta}{2}T_o^2+(k_0-\beta T_o)T_o+\frac{C_1}{r_o}$$

Solve T by using the quadratic formulaï¼š

$$\frac{\beta}{2}T^2+(k_0-\beta T_o)T=-\frac{C_1}{r}+C_2$$

$$\Rightarrow\frac{\beta}{2}T^2+(k_0-\beta T_o)T+(\frac{C_1}{r}+C_2)=0$$

$$T=\frac{\beta T_o-k_0 \pm \sqrt{(k_0-\beta T_o)^2-2\beta(\frac{1}{r}C_1-C_2)}}{\beta}$$

plug in C1 and C2ï¼š

$$(k_0-\beta T_o)^2-2\beta(\frac{1}{r}C_1-C_2)=k_0^2+\beta^2 T_o^2-2k_0\beta T_o-2\beta(\frac{1}{r}C_1-[\frac{\beta}{2}T_o^2+(k_0-\beta T_o)T_o+\frac{C_1}{r_o}])$$

$$=k_0^2+\beta^2 T_o^2-2k_0\beta T_o-2\beta\frac{r_o-r}{rr_o}\frac{r_or_i}{r_i-r_o}(T_i-T_o)(\frac{\beta}{2}(T_i-T_o)+k_0)+2\beta[\frac{\beta}{2}T_o^2+(k_0-\beta T_o)T_o]$$

$$=k_0^2+\beta^2 T_o^2-2k_0\beta T_o+\beta^2 T_o^2+-2k_0\beta T_o-2\beta^2 T_o^2-2\beta\frac{(r_o/r-1)}{(1-r_o/r_i)}(T_i-T_o)(\frac{\beta}{2}(T_i-T_o)+k_0)$$

$$=k_0^2+2\beta\frac{(r_o/r-1)}{(r_o/r_i-1)}(T_i-T_o)(\frac{\beta}{2}(T_i-T_o)+k_0)$$

for Simplifying

$$let\space\space k_m=\frac{1}{2}(k_i+k_0)=\frac{1}{2}[k_0+\beta(T_i-T_o)+k_0]=\frac{\beta}{2}(T_i-T_o)+k_0$$

$$T=T_o+\frac{-k_0 \pm \sqrt{k_0^2+2\beta k_m(T_i-T_o)\frac{(r_o/r-1)}{(r_o/r_i-1)}}}{\beta}\space\space \text{------(5)}$$

We still have a plus-minus sign in front of square root, use boundary condition $T=T_o \space\space\space at\space\space\space r=r_o$ to solve it.

$$\Rightarrow T(r)=T_o+\frac{-k_0 + \sqrt{k_0^2+2\beta k_m(T_i-T_o)\frac{(r_o/r-1)}{(r_o/r_i-1)}}}{\beta}\space\space \text{------(6)}$$

## Discussion of thermal conductivity

In this problem, thermal conductivity is function of temperature, written asï¼š

$$k=k_0+\beta(T-T_o)$$

k-T relationship would be determined by $\beta$ and $k_0$

It's obvious that how $k_0$ affects k, so we could put more emphasis on $\beta$

$$\begin{cases}
\beta <0, &\text{thermal conductivity decreases as temperature increases} \\
\beta >0, &\text{thermal conductivity increases as temperature increases}
\end{cases}$$

In general, thermal conductivities of metal and liquid decrease as temperature increases; thermal conductivities of gas and non-metal increase as temperature increases.

Some of materials have more complicated patterns, like water, aliminum nitride and silicon.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

Ti = 300
To = 50
k0 = 100
ri = 50
ro = 150
B = 0.25
radius = np.linspace(ri,ro,1000)
T = 0*radius
k = 0*radius



fig, ax = plt.subplots(ncols=2, figsize=(14,9)) 

for i, B in enumerate([-0.3,-0.1,-0.05,0.01,0.25,1]):
    km = k0+B/2*(Ti-To)
    for j, r in enumerate(radius):
        T[j] = To+(1/B)*(-k0+np.sqrt(k0**2+2*B*(Ti-To)*km*(ro/r-1)/(ro/ri-1)))
        k[j] = k0+B*(T[j]-To)
    ax[0].plot(radius, k, label='Beta ={}'.format(B))
ax[0].legend(fontsize=16)

ax[0].set_title("Variation of k of different beta with radius", fontsize=16)
ax[0].set_xlabel("Radius r", fontsize=16)
ax[0].set_ylabel("Thermal Conductivity k", fontsize=16)

for i, B in enumerate([-0.3,-0.1,-0.05,0.01,0.25,1]):
    km = k0+B/2*(Ti-To)
    for j, r in enumerate(radius):
        T[j] = To+(1/B)*(-k0+np.sqrt(k0**2+2*B*(Ti-To)*km*(ro/r-1)/(ro/ri-1)))
        k[j] = k0+B*(T[j]-To)
    ax[1].plot(T, k, label='Beta ={}'.format(B))
ax[1].legend(fontsize=16)

ax[1].set_title("Variation of k of different beta with temperature", fontsize=16)
ax[1].set_xlabel("Temperature T", fontsize=16)
ax[1].set_ylabel("Thermal Conductivity k", fontsize=16)

#----------------------------------------------------------------------------------------
B = 0.25
fig, ax = plt.subplots(ncols=2, figsize=(14,9)) 

for i, k0 in enumerate([0.01,0.25,1,10,50,100,200]):
    km = k0+B/2*(Ti-To)
    for j, r in enumerate(radius):
        T[j] = To+(1/B)*(-k0+np.sqrt(k0**2+2*B*(Ti-To)*km*(ro/r-1)/(ro/ri-1)))
        k[j] = k0+B*(T[j]-To)
    ax[0].plot(radius, k, label='k0 ={}'.format(k0))
ax[0].legend(fontsize=16)

ax[0].set_title("Variation of k of different k0 with radius", fontsize=16)
ax[0].set_xlabel("Radius r", fontsize=16)
ax[0].set_ylabel("Thermal Conductivity k", fontsize=16)

for i, k0 in enumerate([0.01,0.25,1,10,50,100,200]):
    km = k0+B/2*(Ti-To)
    for j, r in enumerate(radius):
        T[j] = To+(1/B)*(-k0+np.sqrt(k0**2+2*B*(Ti-To)*km*(ro/r-1)/(ro/ri-1)))
        k[j] = k0+B*(T[j]-To)
    ax[1].plot(T, k, label='k0 ={}'.format(k0))
ax[1].legend(fontsize=16)

ax[1].set_title("Variation of k of different k0 with temperature", fontsize=16)
ax[1].set_xlabel("Temperature T", fontsize=16)
ax[1].set_ylabel("Thermal Conductivity k", fontsize=16)
```

## Discussion of T(r)

This case we have fixed temperature on inner and outer surfaces of sphere, so T(r) converges at r=ri and r=ro.

What beta and k0 change is the profile of T(r) when ri<r<ro.
