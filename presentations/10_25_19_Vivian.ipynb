rectangular T = T (x,y,z,t)
 
$\frac{\partial T}{\partial t}+ v_x \frac{\partial T}{\partial x} + v_y \frac{\partial T}{\partial y}+ v_z \frac{\partial T}{\partial z} = \alpha \bigg[ \frac{\partial^2T}{\partial x^2}+ \frac{\partial^2T}{\partial y^2}+ \frac{\partial^2T}{\partial z^2} \bigg] + \frac{H_v}{\rho C_p}$

Assume density = constant , steady state , no convective forces

therefore we can obtain 

$\frac{k}{\rho C_p} \frac{\partial^2T}{\partial x^2} + \frac{H_v}{k} =0$



after rearrange
$$ \frac{\partial^2T}{\partial x^2} = - \frac{H_v}{k}$$

boundary condition

$T(x=0)= T_0$

$\frac{\partial T}{\partial x}(x=L)=0$


$\dot q= H_v = \dot q  e^ {-\gamma x}$


$$ \frac{\partial^2T}{\partial x^2} = - \frac{ \dot q  e^ {-\gamma x}} {k}$$

$ \frac{\partial^2T}{\partial x^2} = - \int  \frac{ \dot q \gamma e^ {-\gamma x}} {k} dx $

we know that $\frac{\partial T}{\partial x}(x=L)=0$

therefore we can obtain

$\frac{ \dot q e^{-\gamma x}} {k}+ C_1=0 $ 

$C_1= - \frac{ \dot q e^{-\gamma l}} {k \gamma}$

we know that $T(x=0)= T_0$

$T=C_1(x)- \frac{\dot q e^{-\gamma x}} {k \gamma^2}+C_2$

rearrange again

$T_0= -\frac{ \dot q } {k \gamma^2}+C_2$

$C_2= T_0 + \frac{ \dot q}{k \gamma^2}  $

plug in C_1 and C_2

$ T= - \frac{ \dot q e^{-\gamma l}} {k \gamma}x  - \frac{ \dot q e^{-\gamma x}} {k \gamma^2}+T_0+ \frac{\dot q}{k\gamma^2}$

we can obtain the solution

$\frac{ (T-T_0)k \gamma}{\dot q}= -\frac{1}{\gamma}e^{-\gamma x}- e^{-\gamma l} x + \frac{1}{\gamma} $

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import axes3d
import math

x = np.linspace(0,1,101)
y = 0*x
colors = cm.Blues(np.linspace(0, 1, 10))

q = 100
k = 1
l = 1

fig, ax = plt.subplots(figsize=(6,4))


for gamm in (0.01,0.1,1,100):
    T = (q/(k*gamm))* - (1/gamm)* np.exp(-(gamm*x)) - np.exp(- gamm* l)*x +(1/gamm)    
    ax.plot(x, T ,label='\u03B3 = {}'.format(gamm),linewidth=2)


ax.set_xlabel('x')
ax.set_ylabel('T')
#ax.plot(x, T , label=' gamm ', linewidth=15)
#ax.plot(x, T, label='n=0.01')
#ax.plot(x, T, label='n=0')
#ax.plot(x, T, label='n=1')
#ax.plot(x, T, label='n=10')
#ax.plot(x, T, label='n=1000')
ax.legend()
#ax.set_xlim([-2,10])
#ax.set_ylim([-2,10])
```
