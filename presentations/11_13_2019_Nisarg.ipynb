The governing equation for the problem can be given as:

$$\frac{\partial T}{\partial t} = \alpha \frac{\partial^2 T}{\partial x^2}   .....(1)$$

Assuming, 
$\theta_2 = \frac{T}{T_2}$ and $\eta = \frac{x}{g(t)}$

$$\Longrightarrow \frac{\partial T}{\partial x} = T_2 \frac{\partial \theta_2}{\partial x} = T_2 \frac{\partial \theta_2}{\partial \eta} \frac{\partial \eta}{\partial x}$$

$$\frac{\partial^2 T}{\partial x^2} = \frac{\partial}{\partial x} (\frac{\partial T}{\partial x}) = \frac{\partial}{\partial \eta}(T_2 \frac{\partial \theta_2}{\partial \eta} \frac{\partial \eta}{\partial x}) \frac{\partial \eta}{\partial x} = \frac{T_2}{g^2} \frac{\partial^2 \theta_2}{\partial \eta^2} $$

$$\frac{\partial T}{\partial t} = T_2 \frac{\partial \theta}{\partial t} = T_2 \frac{\partial \theta}{\partial \eta} \frac{\partial \eta}{\partial t} = T_2 \frac{\partial \theta}{\partial \eta} (-\frac{x}{g^2}) g' $$

Subtituting the value in eq(1) we get,

$$\Longrightarrow \frac{\partial^2 \theta_2}{\partial \eta^2} = -\frac{gg'}{\alpha}\eta \frac{\partial \theta_2}{\partial \eta}$$

$$\therefore \frac{\partial^2 \theta_2}{\partial \eta^2} + \frac{gg'}{\alpha}\eta \frac{\partial \theta_2}{\partial \eta} = 0 ....(2)$$ 

We have $gg' = (g^2)'$. Solving for it we get $\frac{gg'}{\alpha} = 2$

Substituting the value in eq(2) we get

$$\frac{\partial^2 \theta_2}{\partial \eta^2} + 2\eta \frac{\partial \theta_2}{\partial \eta} = 0......(3)$$

Similarly we get,

$$\frac{\partial^2 \theta_1}{\partial \eta^1} + 2\eta \frac{\partial \theta_1}{\partial \eta} = 0......(4)$$

**Boundary conditions**

1. $x=0 \rightarrow \eta=0$
2. $x= -\infty \rightarrow \eta = -\infty$ and $\theta_2 = 1$
3. $x= \infty \rightarrow \eta = \infty$ and $\theta_1=1$
4. $t=0 \rightarrow \eta = -\infty$

At $x=0$ the fluxes would be equal for surface 1 and 2.

$$\therefore  -k_2 \frac{\partial T_2}{\partial x} = -k_1 \frac{\partial T_1}{\partial x}$$

$$\therefore \frac{k_2 T_2}{g} \frac{\partial \theta_2}{\partial \eta} = \frac{k_1 T_1}{g} \frac{\partial \theta_1}{\partial \eta}$$

$$\Longrightarrow \frac{\partial \theta_2}{\partial \eta} = \frac{k_1 T_1}{k_2 T_2} \frac{\partial \theta_1}{\partial \eta}....(5)$$

Now the solution for equation (3) and (4) can be given as:

$$\frac{\partial \theta_2}{\partial \eta} = a_2 e^{-\eta^2}......(6)$$

and 

$$\frac{\partial \theta_1}{\partial \eta} = a_1 e^{-\eta^2}......(7)$$

Substituting the values of (6) and (7) in eq (5) we get,

$$a_2=\beta a_1.......(8)$$

where $\beta = \frac{k_1 T_1}{k_2 T_2} $

Integrating equations (6) and (7) we get,

$$\theta_2 = a_2 \int_0^x e^{-\eta^2} d\eta + b_2$$

$$\therefore \theta_2 = a_2 (\frac{\sqrt \pi}{2}) erf(\eta) + b_2.......(9)$$

$$\theta_1 = a_1 \int_0^x e^{-\eta^2} d\eta + b_1$$

$$\therefore \theta_1 = a_1 (\frac{\sqrt \pi}{2}) erf(\eta) + b_1........(10)$$

Substituting the boundary conditions, $\eta = -\infty \rightarrow \theta_2 = 1$ and for $\eta = \infty \rightarrow \theta_1 = 1$

$$\Longrightarrow \theta_2 = 1 = -a_2 (\frac{\sqrt \pi}{2}) + b_2......(11)$$

$$\Longrightarrow \theta_1 = 1 = -a_1 (\frac{\sqrt \pi}{2}) + b_1.......(12)$$

At $t=\infty \rightarrow T=T_1$ and at $t=0 \rightarrow T=T_2$

$$\therefore \theta_2 = \frac{T_1 + T_2}{2 T_2} = \frac{1+m}{2}   (m=\frac{T_1}{T_2})$$

$$\theta_1 = \frac{T_1 + T_2}{2 T_1} = \frac{1+m}{2m}$$

$$\Longrightarrow \theta_1 = \frac{\theta_2}{m}........(13) $$

Substituting the values of (11) and (12) in (13) we get,

(At $t=\infty$ $\eta=0$)

$$\Longrightarrow b_2 = mb_1.......(14)$$

Using the values of (8) and (14) in (11) and (12) we get,

$$a_1 = \frac{2(m-1)}{\sqrt \pi (\beta + m)}.......(15)$$

$$b_1 = \frac{\beta + 1}{\beta + m}.......(16)$$

Substituting the values of (15) and (16) in (9) and (10) we get,

$$\frac{T}{T_1} = \frac{T_1 - T_2}{2 T_1} erf(\eta) + \frac{T_1 + T_2}{2 T_1}$$ and

$$\frac{T}{T_2} = \frac{T_1 - T_2}{2 T_2} erf(\eta) + \frac{T_1 + T_2}{2 T_2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

from scipy.special import erf

from matplotlib import animation, rc
from IPython.display import HTML
```

```python
T1 = 40
T2 = 80
n = np.linspace(-10, 10, 100)

T = ((T1-T2)/2)*erf(n) + (T1+T2)/2
```

```python
fig, ax = plt.subplots(figsize = (8,6))
ax.set_title('T vs n')
ax.set_xlabel('n')
ax.set_ylabel('T')
ax.plot(n,T)
ax.grid()
```

```python
a = 10
t = [0.0001, 0.01, 2 , 10, 100, 1000]

c1 = cm.Blues(np.linspace(0.25,1,len(t)))

fig, ax = plt.subplots(figsize = (9,6))
ax.set_title('T vs x')
ax.set_xlabel('x')
ax.set_ylabel('T')
ax.grid()

x = np.linspace(-10,10,100)

for i, ti in enumerate(t):
    T = ((T1-T2)/2)*erf(x/np.sqrt(4*a*ti)) + (T1+T2)/2
    ax.plot(x, T,c = c1[len(t)-1 - i], label = 't = {}'.format(ti))

ax.legend()
plt.show()

```
