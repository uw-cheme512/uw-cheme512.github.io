```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 1
L = 1
q = 100
T0 = 100
x = np.linspace(0, L, 101)

fig, ax = plt.subplots()

c1 = q*L/k
c2 = T0
T = -q*x**2/(2*k) + c1*x + c2
ax.plot(x, T, label=r'$\gamma$ = 0')
for gamma in [0.01, 0.1, 1, 10, 100]:
    
    c1 = -q*np.exp(-gamma*L)/(k*gamma)
    c2 = T0 + q/(k*gamma**2)
    
    T = ((-q/(k*gamma**2))*np.exp(-gamma*x) + c1*x + c2)
    ax.plot(x, T, label=r'$\gamma$ = {}'.format(gamma))
    
ax.legend()
```
