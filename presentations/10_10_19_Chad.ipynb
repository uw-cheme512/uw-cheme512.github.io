```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
Bi1 = 1
H = 1
Bi2 = Bi1*H
Po = 1

x = np.linspace(0, 1, 101)

t1 = (1 - Po*((1/Bi2)+1))/(1 + Bi2 + H)
t2 = Po/Bi2
t3 = Po*(1-x**2)/2
t4 = Bi1*(1-x)*(1+Po*((1/Bi2)+(1/2)))/(1+Bi1+H)
theta = t1+t2+t3+t4
```

```python
fig, ax = plt.subplots(figsize=(7,5))

Pos = np.logspace(-2, 2, 11)
colors = plt.cm.Blues(np.linspace(0,1,11))

Bi1 = 1
H = 1
Bi2 = Bi1*H
#Po = 1

for Po, color in zip(Pos, colors): 
    t1 = (1 - Po*((1/Bi2)+1))/(1 + Bi2 + H)
    t2 = Po/Bi2
    t3 = Po*(1-x**2)/2
    t4 = Bi1*(1-x)*(1+Po*((1/Bi2)+(1/2)))/(1+Bi1+H)
    theta = t1+t2+t3+t4

    ax.plot(x, theta, c=color, label='Po = {}'.format(np.round(Po,2)))

ax.legend()
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel(r'$\Theta$', fontsize=16)
ax.set_title('Bi1 = {}, H = {}'.format(Bi1, H))
```

```python
k = 1
L = 1
T2 = 20
T1 = 10
dT = T2 - T1

Hv = 1
Bi1 = 1
H = 1

fig, ax = plt.subplots(figsize=(7,5))

Hvs = np.logspace(-2, 2, 11)
colors = plt.cm.Blues(np.linspace(0,1,11))

for Hv, color in zip(Hvs, colors): 
    Bi2 = H*Bi1
    c1 = (dT*Bi2 - (H+1)*Hv*L**2/k)/(L*(H - 1 + 2*Bi2))
    c2 = (c1*L*(H - Bi2) + Hv*L**2*(Bi2/2 - 1)/k + Bi2*T2)/Bi2

    T = -Hv*x**2/(2*k) + c1*x + c2
    theta = (T - T1)/dT

    ax.plot(x, theta, c=color, label='Hv = {}'.format(np.round(Hv,2)))

ax.legend()
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel(r'$\Theta$', fontsize=16)
ax.set_title('Bi1 = {}, H = {}'.format(Bi1, H))
```
