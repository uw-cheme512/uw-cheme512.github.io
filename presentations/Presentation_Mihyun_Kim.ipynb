The single reaction arrow indicates that the reaction is irreversible. Then, species A is supplied at x=0, where as species B is supplied at x=L.

The following are given/ assumed to simplify the problem:
$C_i= C_i(x)$ only and $R_{vc}=-R_{va} =-R_{vb} = kC_AC_B$,

where $k$ = second order homogeneous rate constant, and $R_{vC}$= volumetric rate of formation of C.


As stated in the problem statement, if the reaction is fast enough (large k), A and B cannot coexist in the liquid. In order to understand this phenomena in depth, the K, reaction rates, are varied at different values to observe the changes in concentration of A,B within the liquid medium with length L.  
For the concentration profile, following __conservation equation__ is used:
$$D_i\frac{d^2C_i}{dx^2} = -R_{vi}$$

Since, we are focusing on the effects of varying k in terms of the concentration of each species, the following assumptions are made to simplify the problem: steady state, the liquid medium is stagnant, concentration of C is diluted within the system and the concentration profile is 1-D in x-direction.



Assuming $D_A = D_B$ and $C_{A0} = C_{BL}$, conseravtion equation for both A and B becomes:
    $$     D_A\frac{d^2C_A}{dx^2}=-R_{VA}$$
$$      D_A\frac{d^2C_B}{dx^2}=-R_{VB} = -R_{VA}$$

By subtracting both of the equations
  $$     D_A\frac{d^2C_A}{dx^2}-D_A\frac{d^2C_B}{dx^2}=-R_{VA}+-R_{VA}=0$$
  $$\frac{d^2(C_A-C_B)}{dx^2}=0$$

Since 2nd order differential equation is at its simplest form, now we can assume the general solution to this will be:
$$(C_A-C_B) = ax+b$$
The two constants, a and b indicate two boundary conditions are needed for the solution.
__Boundary Conditions.__ Hence:
$$x=L, C_B = C_{BL}$$
which

$$x=0, C = C_{A0}$$
Also, we could assume that 
$$C_{BL} = C_{A0}$$
since the diffusivity terms of species A and B are assumed to be equal, and at each boundaries both concentrations vanished in the opposite wall from the wall they were supplied. 

Using the boundary conditions:
at x=0
$$C_A(0)-C_B(0) = a*0 +b $$
$$b = C_{A0}$$
at x=L
$$C_A(L)-C_B(L) = a*L +b =0 - C_{BL}= -C_{A0}$$
Since $b = C_{A0}$,
$$a = -\frac{2C_{A0}}{L}$$

Plugging a and b back into the linear equation
$$(C_A-C_B) = -\frac{2C_{A0}}{L}x+C_{A0}$$
$$C_A-C_B = C_{A0}(1-\frac{2x}{L})$$

Then,
$$C_B =C_A-C_{A0}(1-\frac{2x}{L})$$

Since we have defined $C_B(x)$, going back to the conservation equation
$$D_A\frac{d^2C_A}{dx^2} =-R_{vA}= kC_AC_B $$
$$D_A\frac{d^2C_A}{dx^2} =-R_{vA}= kC_A(C_A-C_{A0}(1-\frac{2x}{L}))$$
Which our diffusion terms for the species A is defined by the concentration of itself. 

There are two instances when 
$$C_A = C_B$$
which are 1) when they are supplied at the boundary condition assuming their diffusivity constant, $D_A=D_B$ and the fast reaction allows complete irrevesible reaction at some point in within the medium, 2) at the point in the medium when both the concentrations of A and B vanish.
Understanding that, equating the equation for $C_B = C_A$,
then we could see in the case where $D_A= D_B$, and $C_{A0} = C_{BL}$, the reaction is taken place when x=L/2.

#### Dimensionless 
By setting
$$\eta = \frac{x}{L} ; \theta = \frac{C_A}{C_{A0}}$$ 
Using the definition of Damkohler number,
$$Da=\frac{kC_{A0}L^2}{D_A}$$
$$\frac{d^2\theta}{d\eta^2}=Da\theta(\theta+2\eta-1)$$
Hence, imposing the same boundary conditions, the new bowndary condtions are now
$$\theta(0)=1, \theta(1) =0$$
where
$$\theta = \theta(\eta)$$

### Order of Magnitude Study

For the fast reaction where $Da>>1$, the order of magnitude equality is performed to study the behavior of the functions.
let $\delta$=thickness of reaction zone, and $C*$ scale for $C_A$, $C_B$ in the reaction zone.'
Since there are two unknowns, $\delta, C*$, two order of magnitude equations are needed to estimate.

1) From local conservation of A

$$D_A\frac{d^2C_A}{dx^2}= kC_AC_B $$
Since $\Delta C_A $~$ C*$ and $\Delta x$~$\delta$
$D_A\frac{C*}{\delta^2}$~$  kC^2$

$\delta^2$~$\frac{D_A}{kC*}$
Multiplying it by $1/L^2$
$(\frac{\delta}{L})^2$~$\frac{D_A}{kC*L^2}$
$$\frac{D_A}{kC*L^2}=(\frac{D_A}{kC_{A0}L^2})(\frac{C_{A0}}{C*}) = \frac{C_{A0}}{DaC*}$$
Hence,
$$(\frac{\delta}{L})^2\frac{C*}{C_{A0}} $$
equivalent to $$Da^{-1}$$

From the local conservation of A, the order of maginitude studies concludes that the higher initial concentration of A supplied at the boundary increases the reaction rate. 
Also, the overall reaction rate is determined by the ratio $\delta/L$. Meaning, further away the reaction happens from x=0 relative to x=L,the slower the rate of volumetric change in concentraion of A within the system. 

2) From overall conservation of A (Control volume)

$$\int^L_0{D_A\frac{d^2C_A}{dx^2}dx} = \int^L_0{kC_AC_Bdx}$$

$$\int^L_0{D_A\frac{d^2C_A}{dx^2}dx}=  D_A\frac{dC_A}{dx} \Big|_{x=0}^{x=L}=-D_A\frac{dC_A}{dx} \Big|_{x=0}$$
Estimating orders of magnitude:
$-D_A\frac{dC_A}{dx} \Big|_{x=0}$~$D_A(\frac{C_{A0}-C*}{L/2})$~$\frac{D_AC_{A0}}{L}$
Again, dx term is replaced by L/2 as calculated earlier in the first part of the solution.

The right hand side of the conservation equation now becomes 
$\int^L_0{kC_AC_Bdx}$~ $kC*^2\delta$

Equating terms:
$\frac{D_AC_{A0}}{L}$~$kC*^2\delta$

$(\frac{\delta}{L})(\frac{C*}{C_{A0}})^2$~$Da^{-1}$

Since two order of magnitude equations are of the form $x*y^2$
then
$\frac{\delta}{L}\frac{C*}{C_{A0}}$~$Da^{-1/3}$

As $D_a$ indicates the reaction speed, this indicates as L, the length of liquid film increases, reaction rate will slow down, as Da decreases. 
Also, as the initial concentration of the increases, the Da is shown to be at a lower value. This indicates that the reaction rate for this is affected by the initial concentrations of both A and B.
