```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R1 = 1
R2 = 10
fig, ax = plt.subplots(figsize=(6,4))

r = np.linspace(R1, R2, 101)

k = 1
h1 = 2000
h2 = 30
T1 = 100
T2 = 200

for h1 in [0.1, 1, 10, 100, 1000, 10000]:
    frac = ((R2**2*k)/h2 - (1/R2))/((R1**2*k)/h1 + (1/R1))
    c2 = (T2 + T1*frac)/(1 + frac)
    c1 = (c2 - T1)/((R1**2*k)/h1 + (1/R1))

    T = -c1/r + c2

    ax.plot(r, T, label='h1 = {}'.format(h1))

ax.legend()
```

```python
# Book solution
fig, ax = plt.subplots(figsize=(6,4))
h1 = 3
for h2 in [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]:
    theta = ((R2/r) - 1)/((R2/R1) - 1 + (R2/R1)*(k/(R1*h1)) + (k/(R2*h2)))
    T = theta*(T1-T2) + T2

    ax.plot(r, T, label='h2 = {}'.format(h2))

ax.set_ylim([100, 200])
ax.legend()
```
