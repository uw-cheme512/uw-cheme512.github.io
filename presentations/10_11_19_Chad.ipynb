```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1, 201)
Pos = np.linspace(0, 100, 11)
colors = plt.cm.Blues(np.linspace(0,1,11))

fig, ax = plt.subplots(figsize=(7,4))

for Po, color in zip(Pos, colors):
    Tn = np.cos(np.sqrt(Po)*x)/np.cos(np.sqrt(Po))
    ax.plot(x, Tn, c=color, label='Po = {}'.format(np.round(Po,2)))

ax.legend()
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel(r'$\Theta$', fontsize=16)
```

```python
x = np.linspace(0,1, 201)
gams = np.linspace(0, 100, 11)
colors = plt.cm.Blues(np.linspace(0,1,11))

fig, ax = plt.subplots(figsize=(7,4))
L = 1
qd = 100
T0 = 10
k = 1

for gam, color in zip(gams, colors):
    #Tn = qd*(1 - gam*x*np.exp(-gam*L) - np.exp(-gam*x))
    Tn = qd*(1 - gam*x*np.exp(-gam*L) - np.exp(-gam*x))/(k*gam) + T0
    ax.plot(x, Tn, c=color, label=r'$\gamma$ = {}'.format(np.round(gam,2)))

ax.legend()
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel(r'$\Theta$', fontsize=16)
```
