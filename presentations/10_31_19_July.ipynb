#### Problem statement:
A fluid having the temperature $T_i$ flows through a pipe of inner and outer radii $R_i$, $R$.
The inside and outside heat transfer coefficients are $h_i$ and $h_o$, respectively.
The temperature of the outside fluid is $T_o$

$T_o < T_i$ is assumed 

We wish to decrease the heat loss from the inside fluid by insulating the pipe.

Find the heat loss from the inner fluid as a function of the thickness of insulation.
Determine when the maximum heat loss occurs.




#### General equation:
for hollow cylinder that is composed of N layers of materials having different thickness and thermal conductivities. The contact resistance between the layers is negligible.


$$\frac{1}{U_o}=\frac{R_{N+1}/R_1}{h_i}+R_{N+1}\sum_{n=1}\frac{1}{k_n}(\frac{1}{R_n}-\frac{1}{R_{n+1}})+\frac{1}{h_o} \dots (1)$$

where $U_o$ is the overall heat transfer coefficient based on outer surface areas


Ref: V. S. Arpaci, Conduction Heat Transfer (Addison-Wesley, Reading, 1966).

#### Assumptions:
The thermal conductivity of the pipe wall compared with that of the insulation, and the inside heat transfer coefficient compared with the outside coefficient may be assumed to be large.

Because of the relative magnitude of the resistances involved

The inside convective resistance and the conductive resistance of the pipe may be neglected

Hence the inner surface of the insulation assumes the inside fluid temperature Ti approximately.

It follows from appropriate reaangemnet of Eq (1)

$$\frac{1}{U_o}=\frac{R_o}{k}  ln(\frac{R_o}{R})+\frac{1}{h_o} \dots (2) $$

The heat loss from the inside fluid per length $L$ of the pipe:

$$q=U_oA_o(T_i - T_o)$$

where $A_o = 2\pi R_o L$

Rewritten as:

$$q = \frac{2\pi k L(T_i - T_o)}{ln(R_o/R)+(k/hR)/(R_o/R)} \dots (3)$$

*$ln(R_o/R)$: conductive resistance of the insulation

*$(k/hR)/(R_o/R)$:outside convective resistance


from this equation, we can see that when the thickness of the insulation is varied, the first and second terms of the denominator of the right hand side of the equation vary inversely. This suggests the possibility of an extremum for the heat loss from the inside fluid.

rewrite Eq. 3 to:

$$\frac{q}{2\pi k L(T_i - T_o)} = \frac{1}{ln(R_o/R)+(k/hR)/(R_o/R)} \dots (4)$$

find the extremum point by equating to zero the first derivative of Eq. 4 with respect to $R_o/R$:

$$\frac{dq}{d(R_o/R)}=-2\pi kL(T_i-T_o)\frac{1/(R_o/R)-(k/hR)/(R_o/R)^{2}}{(\ln{R_o/R}+(k/hR)/(R_o/R))^{2}}$$

it gives $$R_o/R = k/hR \dots (5)$$    or    $$R_o=k/h \dots (5)$$ when $$\frac{dq}{d(R_o/R)}=0$$


Introducing Eq. 5 to Eq.3, we find that the extremum value of the heat loss from the pipe is:
$$q = \frac{2\pi k L(T_i - T_o)}{1+\ln{(k/hR)}} \dots (6)$$


Furthermore,the second derivative of Eq. 3 evaluated at $R_o/R =k/hR$ yields:

$$\frac{d^2q}{d(R_o/R)^{2}}(R_o/R=k/hR)=-\frac{2\pi Lh^2(T_i - T_o)}{(1+ln(k/hR))^2}<0$$

which indicates that the extremum value of the heat loss given by Eq. 6 corresponds to a maximum.

Surprising result: the heat loss from pipes can be increased by insulation!

It can be further clarified by plotting heat loss from pipes against $R_o/R$ for various values of $1/Bi=k/hR$

$$\frac{q}{2\pi k L(T_i - T_o)} = \frac{1}{ln(R_o/R)+(k/hR)/(R_o/R)} \dots (4)$$
by taking 
$$y= \frac{q}{2\pi k L(T_i - T_o)}$$

$$x=\frac{R_o}{R}$$

$$\frac{k}{hR}=\frac{1}{Bi}$$

Simplifying to:

$$y = \frac{1}{\ln{x}+\frac{1}{Bix}}$$

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import log as ln
```

```python
x = np.linspace(1,5,101)
fig, ax = plt.subplots(figsize=(10,6))

for Bii in [0.8,0.9, 1,1.1,1.2,1.4,1.6,1.8,2,3]:
    y = 1/(np.log(x) + Bii/x)
    ax.plot(x, y, label='1/Bi = {}'.format(np.round(Bii, 4)),)
    
ax.legend()
ax.set_ylim(0,1.5)


plt.xlabel('R0/R')
plt.ylabel('q/(2\pikL(Ti-To))')
plt.show()
```

### *Assume the thickness and material of the insulation is negligible*

Governing equation for cylindrical coordinates:

(Solution manual Eq.19)

$$\rho c \frac{dT}{dt}= \frac{1}{r}\frac{d}{dr}(r\cdot k\frac{dT}{dr})+\frac{1}{r}\frac{d}{d\theta}(k\frac{dT}{d\theta})+\frac{d}{dz}(k\frac{dT}{dz})+q'''$$

simplifying the equation by applying:

$\frac{dT}{dt} = 0$ (steady state)

$\frac{d}{d\theta} = \frac{d}{dz} = 0$ (symetric)

$q''' = 0$ (no internal heating)

now we have:

$$\frac{d^2T}{dr^2}+\frac{1}{r}\frac{dT}{dr}=0$$
