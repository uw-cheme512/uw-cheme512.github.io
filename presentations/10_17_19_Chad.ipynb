```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T0 = 100
T2 = 200
h0 = 0.1
fig, ax = plt.subplots(figsize=(6,4))

ls = np.array([10, 20, 10, 5, 1, 10])
ks = np.array([1, 5, 0.5, 1, 1, 6])
hs = np.array([10, 15, 11, 10, 1, 5])
lb = np.cumsum(ls)

denom = np.sum(ls/ks + 1/hs) + 1/h0

j = 1
x = np.linspace(0, ls[j-1], 100)
theta = (x/ks[j-1] + 1/h0)/denom
ax.plot(x, theta, c='k')


j=2
for j in range(2, 7):
    sums = 0
    sums = np.sum(ls[0:j-1]/ks[0:j-1])

    x = np.linspace(0, ls[j-1], 100)
    theta = (sums + x/ks[j-1] + 1/h0)/denom
    ax.plot(x + lb[j-2], theta, c='k')

#sums = sums + 
```
