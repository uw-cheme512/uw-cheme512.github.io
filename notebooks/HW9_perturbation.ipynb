**Case 1:** $\epsilon^n$

We need to find a function of the form:

$$\ln{(x + \epsilon)} = \sum_{n=0}^{\infty}{y_n(x)\epsilon^n} $$

For the first term:

$$y_0 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon)}}{1}} = \ln{x} $$

And the second term:

$$y_1 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon) - \ln{x}}}{\epsilon}} = \lim_{\epsilon \rightarrow 0}{\frac{\frac{1}{x+\epsilon}}{1}} = \frac{1}{x} $$

Thus:

$$\ln{(x + \epsilon)} = \ln{x} + \epsilon \frac{1}{x} - \frac{\epsilon^2}{2}\frac{1}{x^2} + O(\epsilon^3) $$

**Case 2:** $\sin^n{\epsilon}$

We need to find a function of the form:

$$\ln{(x + \epsilon)} = \sum_{n=0}^{\infty}{y_n(x)\sin^n{\epsilon}} $$

For the first term:

$$y_0 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon)}}{1}} = \ln{x} $$

And the second term:

$$y_1 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon) - \ln{x}}}{\sin{\epsilon}}} = \lim_{\epsilon \rightarrow 0}{\frac{\frac{1}{x+\epsilon}}{\cos{\epsilon}}} = \frac{1}{x} $$

And the third term:

$$y_1 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon) - \sin{\epsilon} \frac{1}{x} - \ln{x}}}{\sin^2{\epsilon}}} = \lim_{\epsilon \rightarrow 0}{\frac{\frac{1}{x+\epsilon} - \cos{\epsilon}\frac{1}{x} }{2 \sin{\epsilon} \cos{\epsilon}}} = \lim_{\epsilon \rightarrow 0}{\frac{- \frac{1}{(x+\epsilon)^2} + \sin{\epsilon}\frac{1}{x} }{2 (\cos^2{\epsilon} - \sin^2{\epsilon}) }} = -\frac{1}{2 x^2} $$

Thus:

$$\ln{(x + \epsilon)} = \ln{x} + \sin{\epsilon} \frac{1}{x} - \frac{\sin^2{\epsilon}}{2}\frac{1}{x^2} + O(\epsilon^3) $$

**Case 3:** $[\ln{(1 + \epsilon)}]^n$

We need to find a function of the form:

$$\ln{(x + \epsilon)} = \sum_{n=0}^{\infty}{y_n(x)[\ln{(1 + \epsilon)}]^n} $$

For the first term:

$$y_0 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon)}}{1}} = \ln{x} $$

And the second term:

$$y_1 = \lim_{\epsilon \rightarrow 0}{\frac{\ln{(x + \epsilon) - \ln{x}}}{\ln{(1 + \epsilon)}}} = \lim_{\epsilon \rightarrow 0}{\frac{\frac{1}{x+\epsilon}}{\frac{1}{1 + \epsilon}}} = \frac{1}{x} $$

Thus:

$$\ln{(x + \epsilon)} = \ln{x} + \ln{(1 + \epsilon)} \frac{1}{x} - [\ln{(1 + \epsilon)}]^2\frac{1}{2 x^2} + O(\epsilon^3) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
x = np.linspace(0.1, 1, 101)

fig, ax = plt.subplots(ncols=2, figsize=(12, 5))
colors = cm.Blues(np.linspace(0, 1, 15))
colors2 = cm.Reds(np.linspace(0, 1, 15))

for i, eps in enumerate(np.logspace(-4, 0, 15)):
    y = np.log(x) + eps/x - eps**2/(2*x**2)
    ax[0].plot(x, y, c=colors[i], label=r'$\epsilon$ = {}'.format(np.round(eps, 4)))
    ax[1].plot(x, np.log(x+eps), c=colors2[i], label=r'$\epsilon$ = {}'.format(np.round(eps, 4)))
ax[0].legend(loc=(1.02,0.01))
ax[0].set_ylim([-4, 1])
ax[1].legend(loc=(1.02,0.01))
ax[1].set_ylim([-4, 1])
fig.subplots_adjust(wspace=0.55)
```

#### Problem 3 (Morbidelli 9.13)

Use the method of matched asymptotic expansions to find the zeroth-order approximation for the solution to the equation

$$\epsilon \frac{d^2 y}{dx^2} + \beta \frac{dy}{dx} = 1, \beta > 0 $$

with BCs

$$y = \alpha, x=0 $$
$$y=1, x=1 $$

Compare the result with the zeroth-order expansion of the analytical solution.

**Analytical solution**

$$y_h = C_1 + C_2 e^{-\beta x/ \epsilon} $$
$$\frac{d y_h}{dx} = - C_2 \frac{\beta}{\epsilon} e^{-\beta x/ \epsilon} $$
$$\frac{d^2 y_h}{dx^2} = C_2 \frac{\beta^2}{\epsilon^2} e^{-\beta x/ \epsilon} $$

Our guess for the particular solution is:

$$y_p = A_1 t $$

For a full solution of form:

$$y = C_1 + C_2 e^{-\beta x/ \epsilon} + A_1 x$$

Plugging back into the original ODE gives $A_1 = 1/\beta$.

Using our initial conditions to solve for the other coefficients:

$$C_2 = \frac{1 - \alpha + \frac{1}{\beta}}{e^{-\beta/ \epsilon} -1} $$
$$C_1 = \frac{\alpha e^{-\beta / \epsilon} - 1 + \frac{1}{\beta}}{e^{-\beta / \epsilon} -1} $$

Let's try plotting the solution for various values of $\beta$ and $\alpha$.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python

beta = 1
alpha = 1
eps = 0.1

colors = cm.Blues(np.linspace(0, 1, 20))
fig, ax = plt.subplots()
for i, eps in enumerate(np.logspace(0, -3, 20)):
    c1 = (alpha*np.exp(-beta/eps) - 1 + (1/beta))/(np.exp(-beta/eps) -1)
    c2 = (1 - alpha - (1/beta))/(np.exp(-beta/eps)-1)

    x = np.linspace(0, 1, 101)
    y = c1 + c2*np.exp(-beta*x/eps) + x/beta

    ax.plot(x, y, c=colors[i])
```

We now know that the boundary layer is located close to 0, and the zeroth order term of the outer expansion is likely going to be $y=x$, at least in the case of $\beta=1$. Let's see if we can find that now with perturbation theory:

We will cheat with some knowledge from the book, that in the case of a very similar ODE, the transformation for the inner expansion is $s = x/\epsilon$. Doing some term-matching for the case of the outer expansion, we get:

$$\frac{dy_0^o}{dx} = \frac{1}{\beta}, y_0^o(1)=1$$

Solving gives

$$y_0^o = \frac{1}{\beta}(x-1) + 1 $$

For the inner expansion, the ODE becomes:

$$\frac{d^2y}{ds^2} + \beta \frac{dy}{ds} = \epsilon, y=\alpha, s=0 $$

Matching zero-order terms:

$$\frac{d^2y_0^i}{ds^2} + \beta \frac{dy_0^i}{ds} = 0, y_0^i(0) = \alpha $$

```python
beta = 1
alpha = 1
eps = 0.1

colors = cm.Blues(np.linspace(0, 1, 20))
fig, ax = plt.subplots(ncols=2, figsize=(12,5))
for i, eps in enumerate(np.logspace(0, -3, 20)):
    c1 = (alpha*np.exp(-beta/eps) - 1 + (1/beta))/(np.exp(-beta/eps) -1)
    c2 = (1 - alpha - (1/beta))/(np.exp(-beta/eps)-1)

    x = np.linspace(0, 1, 101)
    s = x/eps
    y = c1 + c2*np.exp(-beta*x/eps) + x/beta
    yo = (x-1)/beta + 1
    ci = alpha + (1/beta) - 1
    yi = alpha + ci*(np.exp(-beta*s)-1)

    ax[0].plot(x, y, c=colors[i])
    ax[1].plot(x, yo, c=colors[i])
    ax[1].plot(x, yi, c=colors[i])
```

#### Problem 5 (Morbidelli 9.20 modified)

Find the approximate solution of the third-order differential equation

$$\epsilon \frac{d^3y}{dx^3} - x^3y = 0$$

using the WKB method

**Plot the solution with varying values of $\epsilon$**.

We are going to define the solution in terms of an exponential power series:

$$y = e^{\sum_{n=0}^{\infty}{\delta^{n-1}S_n(x)}} $$

Taking derivatives:

$$\frac{dy}{dx} = \bigg( \frac{1}{\delta}\frac{dS_0}{dx} + \frac{dS_1}{dx} + \delta \frac{dS_2}{dx} + ... \bigg)y$$

$$\frac{d^2y}{dx^2} = \bigg( \frac{1}{\delta}\frac{dS_0}{dx} + \frac{dS_1}{dx} + \delta \frac{dS_2}{dx} + ... \bigg)^2y + \bigg( \frac{1}{\delta}\frac{d^2S_0}{dx^2} + \frac{d^2S_1}{dx^2} + \delta \frac{d^2S_2}{dx^2} + ... \bigg)y$$

$$\frac{d^3y}{dx^3} = \bigg( \frac{1}{\delta}\frac{dS_0}{dx} + \frac{dS_1}{dx} + \delta \frac{dS_2}{dx} + ... \bigg)^3y + 3 \bigg( \frac{1}{\delta}\frac{dS_0}{dx} + \frac{dS_1}{dx} + \delta \frac{dS_2}{dx} + ... \bigg) \bigg( \frac{1}{\delta}\frac{d^2S_0}{dx^2} + \frac{d^2S_1}{dx^2} + \delta \frac{d^2S_2}{dx^2} + ... \bigg)y + \bigg( \frac{1}{\delta}\frac{d^3S_0}{dx^3} + \frac{d^3S_1}{dx^3} + \delta \frac{d^2S_3}{dx^3} + ... \bigg)y$$

Substition into the equation gives:

$$\epsilon \bigg( \frac{1}{\delta}\frac{dS_0}{dx} + \frac{dS_1}{dx} + \delta \frac{dS_2}{dx} + ... \bigg)^3 + 3 \epsilon \bigg( \frac{1}{\delta}\frac{dS_0}{dx} + \frac{dS_1}{dx} + \delta \frac{dS_2}{dx} + ... \bigg) \bigg( \frac{1}{\delta}\frac{d^2S_0}{dx^2} + \frac{d^2S_1}{dx^2} + \delta \frac{d^2S_2}{dx^2} + ... \bigg) + \epsilon \bigg( \frac{1}{\delta}\frac{d^3S_0}{dx^3} + \frac{d^3S_1}{dx^3} + \delta \frac{d^2S_3}{dx^3} + ... \bigg) - x^3 = 0 $$


Setting $\delta = (\epsilon)^{1/3}$ (which gives $\epsilon/\delta = \epsilon^{2/3}$ and $\epsilon/\delta^2 = \epsilon^{1/3}$) and combining like terms:

$$\bigg[ \bigg( \frac{dS_0}{dx} \bigg)^3 - x^3  \bigg] + \epsilon^{1/3} \bigg[ 3 \bigg( \frac{dS_0}{dx} \bigg)^2 \bigg( \frac{dS_1}{dx} \bigg) + 3 \bigg( \frac{dS_0}{dx} \bigg) \bigg( \frac{d^2S_0}{dx^2} \bigg) \bigg] + \epsilon^{2/3} \bigg[ 3 \bigg( \frac{dS_0}{dx} \bigg)^2 \bigg( \frac{dS_2}{dx} \bigg) + 3 \bigg( \frac{dS_0}{dx} \bigg) \bigg( \frac{dS_1}{dx} \bigg)^2  + 3 \bigg( \frac{d^2S_0}{dx^2} \bigg) \bigg( \frac{dS_1}{dx} \bigg) + 3 \bigg( \frac{dS_0}{dx} \bigg) \bigg( \frac{d^2S_0}{d^2x} \bigg) + \bigg( \frac{d^3S_0}{dx^3} \bigg) \bigg]$$

Each term should equal zero:

$$O(\epsilon^0): \bigg( \frac{dS_0}{dx} \bigg)^3 - x^3 = 0 $$
$$S_0 = \frac{1}{2}x^2 $$

$$O(\epsilon^{1/3}): 3 \bigg( \frac{dS_0}{dx} \bigg)^2 \bigg( \frac{dS_1}{dx} \bigg) + 3 \bigg( \frac{dS_0}{dx} \bigg) \bigg( \frac{d^2S_0}{dx^2} \bigg) = 0$$
$$S_1 = -\ln{x} $$

$$O(\epsilon^{2/3}): 3 \bigg( \frac{dS_0}{dx} \bigg)^2 \bigg( \frac{dS_2}{dx} \bigg) + 3 \bigg( \frac{dS_0}{dx} \bigg) \bigg( \frac{dS_1}{dx} \bigg)^2  + 3 \bigg( \frac{d^2S_0}{dx^2} \bigg) \bigg( \frac{dS_1}{dx} \bigg) + 3 \bigg( \frac{dS_0}{dx} \bigg) \bigg( \frac{d^2S_0}{d^2x} \bigg) + \bigg( \frac{d^3S_0}{dx^3} \bigg) = 0$$
$$S_2 = -\ln{x} $$

This gives the final expression:
    
$$y = C_1 e^{\frac{1}{2 \delta}x^2 - (1 + \delta) \ln{x}} $$

```python
eps = 0.01
colors = cm.Blues(np.linspace(0,1,15))
fig, ax = plt.subplots()
for i, eps in enumerate(np.logspace(-4, 0.7, 15)):
    delta = eps**(1/3)

    x = np.linspace(0.1, 0.99, 101)
    #x = 0.85-x1
    term = 0.5*x**2/delta - np.log(x) - delta*np.log(x)
    y = np.exp(term)

    ax.plot(x, y, c=colors[i], label=r'\epsilon = {}'.format(np.round(eps, 4)))
ax.plot(x, x**(-3), c='k', label=r'\epsilon = 0')
ax.set_ylim([-1, 100])
ax.legend(loc=(1.05,0.01))
```
